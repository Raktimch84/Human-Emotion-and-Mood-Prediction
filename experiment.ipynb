```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
data=pd.read_csv('tweet_emotions.csv')
```

```python
data.groupby('sentiment').describe()
```

```python
data.drop(['tweet_id'],axis=1)
```

```python
data['emotion']=data['sentiment'].apply(lambda x: 1 if x=='enthusiasm' else 2 if x=='love' else 3 if x=='fun' else 4 if x=='happiness' else 5 if x=='relief' else 6 if x=='sadness' else 7 if x=='worry' else 8 if x=='boredom' else 9 if x=='neutral' else 10 if x=='empty' else 11 if x=='anger' else 12 if x=='hate' else 13)
```

```python
data.head(15)
```

```python
x_train,x_test,y_train,y_test= train_test_split(data['content'],data['emotion'],test_size=0.20,random_state=2)
```

```python
vector=CountVectorizer()
```

```python
x_count=vector.fit_transform(x_train)
```

```python
mod=MultinomialNB()
```

```python
mod.fit(x_count,y_train)
```

```python
text=["guys we did it we won"]
t_c=vector.transform(text)
if(mod.predict(t_c))<=5:
     print("HAPPY")
elif mod.predict(t_c)>5 and mod.predict(t_c)<=10 :
     print("SAD")
else:
     print("ANGRY")
mod.predict(t_c)
```

```python
count=vector.transform(x_test)
mod.score(count,y_test)
```
